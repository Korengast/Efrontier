```python
import pandas as pd
import numpy as np
```

```python
file_path = 'predictions/5M_30M/'
file_name = 'RandomForest_1000_BNBUSDT.csv'
res_df = pd.read_csv(file_path + file_name)
```

```python
avg_inc_pred = np.mean(res_df[res_df['y_bins'] > 0]['predictions'])
avg_dec_pred = np.mean(res_df[res_df['y_bins'] < 0]['predictions'])
```

```python
print(avg_inc_pred)
print(avg_dec_pred)
```

```python
avg_inc_bins = np.mean(res_df[res_df['predictions'] > avg_inc_pred]['y_bins'])
avg_dec_bins = np.mean(res_df[res_df['predictions'] < avg_dec_pred]['y_bins'])
```

```python
print(avg_inc_bins)
print(avg_dec_bins)
```

```python
features_imp = pd.DataFrame({'feature': res_df.columns,
                'importance': res_df.iloc[-1].values*100})
```
