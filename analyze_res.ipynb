```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
file_path = 'predictions/5M_30M/'
file_name = 'RandomForest_100_NEOUSDT.csv'
res_df = pd.read_csv(file_path + file_name)
```

```python

start_time = datetime.utcfromtimestamp(int(res_df['timestamp'].iloc[0]/1000))
end_time = datetime.utcfromtimestamp(int(res_df['timestamp'].iloc[-2]/1000))
print('Results file starts at: ' + str(start_time))
print('Results file ends at: ' + str(end_time))
```

```python
avg_inc_pred = np.mean(res_df[res_df['y_bins'] > 0]['predictions'])
avg_dec_pred = np.mean(res_df[res_df['y_bins'] < 0]['predictions'])
avg_zero_pred = np.mean(res_df[res_df['y_bins'] == 0]['predictions'])
```

```python
print(avg_inc_pred)
print(avg_zero_pred)
print(avg_dec_pred)
```

```python
avg_inc_bins = np.mean(res_df[res_df['predictions'] > avg_inc_pred]['y_bins'])
avg_dec_bins = np.mean(res_df[res_df['predictions'] < avg_dec_pred]['y_bins'])
```

```python
print(avg_inc_bins)
print(avg_dec_bins)
```

```python
print(np.mean(res_df[res_df['predictions'] > avg_inc_pred]['y%']))
print(np.sum(res_df[res_df['predictions'] < avg_dec_pred]['y%']))
```
